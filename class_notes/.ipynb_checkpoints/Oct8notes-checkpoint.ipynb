```python
earthdf = pd.read_csv('http://www.ldeo.columbia.edu/~rpa/usgs_earthquakes_2014.csv',
                      index_col = 'id', parse_dates=['time'])
```

```python
earthdf
```

```python
earthdf['country'] = earthdf.place.str.split(', ').str[-1]
```

```python
earthdf = earthdf[earthdf.mag >=4]
```

```python
gb = earthdf.groupby(earthdf.country)
```

```python
len(gb)
```

```python
gb.size()
```

```python
gb.count()
```
