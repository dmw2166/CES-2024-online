```python
import urllib

header_url = 'ftp://ftp.ncdc.noaa.gov/pub/data/uscrn/products/daily01/HEADERS.txt'
with urllib.request.urlopen(header_url) as response:
    data = response.read().decode('utf-8')
lines = data.split('\n')
headers = lines[1].split(' ')

ftp_base = 'ftp://ftp.ncdc.noaa.gov/pub/data/uscrn/products/daily01/'
dframes = []
for year in range(2016, 2019):
    data_url = f'{year}/CRND0103-{year}-NY_Millbrook_3_W.txt'               
    df = pd.read_csv(ftp_base + data_url, parse_dates=[1],
                     names=headers, header=None, sep='\s+',
                     na_values=[-9999.0, -99.0])
    dframes.append(df)

df = pd.concat(dframes)  #concatenates files by dates here, axis=0 (rows by default)
df = df.set_index('LST_DATE')
```

```python
df
```

```python
df.T_DAILY_MEAN.plot()
```

```python
df.groupby(df.index.month).mean(numeric_only=True)
```

```python
monthly_clim = df.groupby(df.index.month).aggregate({'T_DAILY_MEAN':'mean',
                                                     'T_DAILY_MAX': 'max', 
                                                     'T_DAILY_MIN':'min'})
monthly_clim
```

```python
monthly_clim.plot()
```

```python
daily_clim = df.groupby(df.index.dayofyear).aggregate({'T_DAILY_MEAN':'mean',
                                                     'T_DAILY_MAX': 'max', 
                                                     'T_DAILY_MIN':'min'})
daily_clim.plot(marker='^')
```

```python
anomaly = df['T_DAILY_MEAN'].groupby(df.index.month).transform(standardize)
anomaly.plot()
```
