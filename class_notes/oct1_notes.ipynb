```python
x = np.linspace(-np.pi,np.pi,100)
y = np.cos(x)
z = np.sin(6*x)
```

```python
fig,ax = plt.subplots()
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(ncols=2,figsize=(8,4))
axes[0].plot(x,y)
axes[1].plot(x,z)
```

```python
fig,axes = plt.subplots(ncols=2,figsize=(8,4))
ax0,ax1 = axes
ax0.plot(x,y)
ax1.plot(x,z)
```

```python
fig,axes = plt.subplots(ncols=2,figsize=(12,4))
ax0,ax1 = axes
ax0.plot(x,y)
ax1.plot(x,z)
ax0.set_xlabel('x values')
ax0.set_ylabel('y values')
ax0.set_title('x versus y')

ax1.set_xlabel('x values')
ax1.set_ylabel('z values')
ax1.set_title('x versus z')
```

```python
fig, ax = plt.subplots()
ax.plot(x,y,x,z) #x1, y1, x2,y2
```

```python
fig,axes = plt.subplots(ncols=2,figsize=(12,4))
ax0,ax1 = axes
ax0.plot(x,y,linestyle='-',linewidth=1, color=(1,0.4,0.2),marker='.')
ax1.plot(x,z,linestyle='-.',color='mediumvioletred',marker='^',markersize=10,
        markeredgecolor='k')
ax0.set_xlabel('x values')
ax0.set_ylabel('y values')
ax0.set_title('x versus y')

ax1.set_xlabel('x values')
ax1.set_ylabel('z values')
ax1.set_title('x versus z')
```

```python
fig,axes = plt.subplots(ncols=2,figsize=(12,4))
ax0,ax1 = axes
ax0.plot(x,y,linestyle='-',linewidth=1, color=(1,0.4,0.2),marker='.')
ax1.plot(x,z,linestyle='-.',color='mediumvioletred',marker='^',markersize=10,
        markeredgecolor='k')
ax0.set_xlabel('x values')
ax0.set_ylabel('y values')
ax0.set_title('x versus y')
ax0.set_xticks(np.array([-np.pi,0,np.pi]))
ax0.set_xticks(np.arange(-np.pi,np.pi,0.3),minor=True)
ax0.set_xticklabels(['$-\pi$','0','$\pi$'])
ax0.grid(which='major',linewidth=0.5)
ax0.grid(which='minor',linewidth=0.25)


ax1.set_xlabel('x values')
ax1.set_ylabel('z values')
ax1.set_title('x versus z')
```

```python
t = np.arange(0,10,0.1)
u = np.exp(t)
v = np.sin(2 * np.pi * t) 
```

```python
u
```

```python
v
```

```python
fig,ax = plt.subplots()
ax.plot(t,u,color='tab:red')
ax.tick_params(axis='y',labelcolor='tab:red')
ax.set_ylabel('Left axis', color='tab:red')

ax2 = ax.twinx()
ax2.plot(t,v,color='tab:blue')
ax2.tick_params(axis='y',labelcolor='tab:blue')
ax2.set_ylabel('Right axis',color='tab:blue')
```

```python
labels = ['first','second','third']
values = [10,5,30]
```

```python
fig,axes = plt.subplots(ncols=2,figsize=(10,5))
axes[0].bar(labels,values)
axes[1].barh(labels,values)
```

```python
labels = ['first','second','third']
values = [10,5,30]
values2 = [20,10,60]
```

```python
fig,ax = plt.subplots()

locs = np.arange(len(labels))
width = 0.2
offset = 0.1

ax.bar(locs-offset,values,width, label='values')
ax.bar(locs+offset, values2,width, label='values2')
ax.legend()
ax.set_xticks(locs,labels)
```
